```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the two CSV files into DataFrames
df_LP_canada = pd.read_csv('data/df_LP_Canada.csv')
df_LP_US = pd.read_csv('data/df_LP_US.csv')

# Concatenate the two DataFrames vertically
data = pd.concat([df_LP_canada, df_LP_US], ignore_index=True)
```

```python
data
```

```python
#drop unnamed
data = data.drop(columns=['Unnamed: 0'])
```

```python
# Display the unique values of the 'category' column
unique_categories1 = data['category1_code'].unique()

# Print the unique categories
print(sorted(unique_categories1))
```

```python
# Display the unique values of the 'category' column
unique_categories2 = data['category2_code'].unique()

# Print the unique categories
print(sorted(unique_categories2))
```

```python
unique_categories = data[['category1_code', 'category2_code']].sort_values(by='category1_code').drop_duplicates().values
unique_categories
```

```python
# category1 to group mapping LP Canada
category1_to_group_mapping = {
    'MEN': ['MAN','FOR-HIM','GIFTS-FOR-HIM'],
    'WOMEN': ['WOMAN'],
    'KIDS': ['KIDS'],
    'OTHER': ['AT-HOME'],
    'GENDER NEUTRAL': ['GIFT-GUIDE','GIFTING-IDEA','EXTRA-BAG-AND-EXTRA-CASE','POLO']
}

# Function to assign group based on category1_code
def assign_group_category1(category):
    for label, associated_categories in category1_to_group_mapping.items():
        if category in associated_categories:
            return label
    return None

# Add 'Group' column to the DataFrame df_LP_canada_US
data['Group'] = data['category1_code'].apply(assign_group_category1)
```

```python
# category1 to category mapping LP Canada
category1_to_category_mapping = {
    'SHOES': ['SHOES'],
    'KIDS': ['KIDS'],
    'DECOR/HOME':['AT-HOME'],
    'GIFTS': ['GIFT-GUIDE','GIFTING-IDEA','GIFTS-FOR-HIM'],
    'BAGS':['EXTRA-BAG-AND-EXTRA-CASE'],
    'OTHER': [],
    'CLOTHING': ['TOPS']
}

# Function to assign category based on category1_code
def assign_category_category1(category):
    for label, associated_categories in category1_to_category_mapping.items():
        if category in associated_categories:
            return label
    return None

# Add 'Category' column to the DataFrame df_LP_canada_US
data['Category'] = data['category1_code'].apply(assign_category_category1)
```

```python
# category2 to group mapping LP Canada
category2_to_group_mapping = {
    'MEN': ['MAN',],
    'WOMEN': ['WOMAN'],
    'KIDS': ['KIDS']
}

# Function to assign group based on category2_code
def assign_group_category2(row):
    if pd.isna(row['Group']):  # Check if 'Group' column is NaN or None
        for label, associated_categories in category2_to_group_mapping.items():
            if row['category2_code'] in associated_categories:
                return label
    return row['Group']

# Apply the function only where 'Group' column is None
data['Group'] = data.apply(assign_group_category2, axis=1)
```

```python
# category2 to category mapping LP Canada
category2_to_category_mapping = {
    'KIDS': ['KIDS'],
    'SHOES': ['SHOES',],
    'CLOTHING': ['VESTS','TOP-AND-T-SHIRTS','THE-GIFT-OF-KINGS','VICUNA','BLAZERS','SHAWL-DRESSING','SKIRTS','SHIRTS-AND-TOPS','DRESSES','BEACHWEAR','COATS-AND-CAPES','TROUSERS-AND-SHORTS','JACKETS-AND-GILETS','TROUSERS-AND-BERMUDAS','KNITWEAR','OUTERWEAR-JACKETS','BLAZERS','SHIRTS','BABY-CASHMERE','COATS-AND-TRENCH','VICUNA','POLO-AND-T-SHIRTS','SWIMWEAR'],
    'ACCESSORIES':['ACCESSORIES','ROYAL-COLLEGE-SCARF-FAL3110'],
    'BAGS':['LEATHER-GOODS']
}
 

# Function to assign group based on category2_code
def assign_category_category2(row):
    if pd.isna(row['Category']):  # Check if 'Group' column is NaN or None
        for label, associated_categories in category2_to_category_mapping.items():
            if row['category2_code'] in associated_categories:
                return label
    return row['Category']

# Apply the function only where 'Group' column is None
data['Category'] = data.apply(assign_category_category2, axis=1)
```

```python
category_None_data = data['Category'].isnull()
data.loc[category_None_data & data['itemurl'].str.contains(r'shirt|crew-neck|cocooning-barrette|garden-oversized|oversized-foulard|darwin-snood|jacket', case=False), 'Category'] = 'CLOTHING'
data.loc[category_None_data & data['itemurl'].str.contains(r'scarf|stole|scrunchie|shawl|quadrilatero', case=False), 'Category'] = 'ACCESSORIES'
data.loc[category_None_data & data['itemurl'].str.contains(r'sneakers|slippers', case=False), 'Category'] = 'SHOES'

```

```python
# Filter rows where either 'groups' or 'category' column is NaN
nan_rows = data[data['Group'].isna() | data['Category'].isna()]


nan_rows
```

```python
# Write the DataFrame back to a CSV file
data.to_csv('data/df_LP_canada_US.csv', index=False)
```
