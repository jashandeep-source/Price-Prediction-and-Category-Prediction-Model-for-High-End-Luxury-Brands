```python
import pandas as pd
```

```python
#reading the file
df_Gucci_Canada = pd.read_csv('data/df_Gucci_Canada.csv')
df_Gucci_US = pd.read_csv('data/df_Gucci_US.csv')
```

```python
#drop unnamed
df_Gucci_Canada = df_Gucci_Canada.drop(columns=['Unnamed: 0'])
df_Gucci_US = df_Gucci_US.drop(columns=['Unnamed: 0'])
```

```python
# Obtain the unique values of Category_code_1
unique_category_code_1_Gucci_US = df_Gucci_US['category1_code'].unique()

print(unique_category_code_1_Gucci_US)

unique_category_code_1_Gucci_Canada = df_Gucci_Canada['category1_code'].unique()

print(unique_category_code_1_Gucci_Canada)
```

```python
# Obtain the unique values of Category_code_2
unique_category_code_2_Gucci_US = df_Gucci_US['category2_code'].unique()

print(unique_category_code_2_Gucci_US)

unique_category_code_2_Gucci_Canada = df_Gucci_Canada['category2_code'].unique()

print(unique_category_code_2_Gucci_Canada)
```

```python
#Transformation for Category1_code
df_Gucci_US['Category'] = 'OTHER'

df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'bags|backpacks|GUCCI-BAMBOO-1947', case=False) , 'Category'] = 'BAGS'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'accessories|-acc-|jewelry|leather-goods', case=False), 'Category'] = 'ACCESSORIES'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'travel|monogram', case=False), 'Category'] = 'TRAVEL'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'shoes', case=False), 'Category'] = 'SHOES'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'READY-TO-WEAR', case=False)|df_Gucci_US['category2_code'].str.contains(r'READY-TO-WEAR', case=False)| df_Gucci_US['category3_code'].str.contains(r'READY-TO-WEAR', case=False),'Category'] = 'CLOTHING'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'watches', case=False), 'Category'] = 'WATCHES'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'make-up', case=False), 'Category'] = 'MAKEUP'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'girls', case=False), 'Category'] = 'GIRLS'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'boy', case=False), 'Category'] = 'BOYS'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'baby', case=False), 'Category'] = 'KIDS'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'fragrances', case=False), 'Category'] = 'FRAGRANCES'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'decor', case=False), 'Category'] = 'DECOR/HOME'
df_Gucci_US.loc[df_Gucci_US['category1_code'].str.contains(r'gift', case=False), 'Category'] = 'GIFTS'


df_Gucci_Canada['Category'] = 'OTHER'

df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'bags|backpacks|GUCCI-BAMBOO-1947', case=False) , 'Category'] = 'BAGS'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'accessories|-acc-|jewelry|leather-goods', case=False), 'Category'] = 'ACCESSORIES'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'travel|monogram', case=False), 'Category'] = 'TRAVEL'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'shoes', case=False), 'Category'] = 'SHOES'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'READY-TO-WEAR', case=False)|df_Gucci_Canada['category2_code'].str.contains(r'READY-TO-WEAR', case=False)| df_Gucci_Canada['category3_code'].str.contains(r'READY-TO-WEAR', case=False),'Category'] = 'CLOTHING'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'watches', case=False), 'Category'] = 'WATCHES'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'make-up', case=False), 'Category'] = 'MAKEUP'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'girls', case=False), 'Category'] = 'GIRLS'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'boy', case=False), 'Category'] = 'BOYS'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'baby', case=False), 'Category'] = 'KIDS'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'fragrances', case=False), 'Category'] = 'FRAGRANCES'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'decor', case=False), 'Category'] = 'DECOR/HOME'
df_Gucci_Canada.loc[df_Gucci_Canada['category1_code'].str.contains(r'gift', case=False), 'Category'] = 'GIFTS'

```

```python
#count the records in each group
category_count_Gucci_US = df_Gucci_US['Category'].value_counts()
print("Category count in Gucci_US")
print(category_count_Gucci_US)
category_count_Gucci_Canada = df_Gucci_Canada['Category'].value_counts()
print("Category count in Gucci_Canada")
print(category_count_Gucci_Canada)
```

```python
#Transformation for Category2_code to get the Group
df_Gucci_US['Group'] = 'OTHER'

df_Gucci_US.loc[df_Gucci_US['category2_code'].str.contains('MEN', case=False)|df_Gucci_US['category1_code'].str.contains('MEN', case=False), 'Group'] = 'MEN'
df_Gucci_US.loc[df_Gucci_US['category2_code'].str.contains('WOMEN', case=False)|df_Gucci_US['category1_code'].str.contains('MEN', case=False), 'Group'] = 'WOMEN'
df_Gucci_US.loc[df_Gucci_US['category2_code'].str.contains('BOYS', case=False), 'Group'] = 'BOYS'
df_Gucci_US.loc[df_Gucci_US['category2_code'].str.contains('GIRLS', case=False), 'Group'] = 'GIRLS'
df_Gucci_US.loc[df_Gucci_US['category2_code'].str.contains(r'BABY|BABIES|TODDLER', case=False), 'Group'] = 'KIDS'

# Adding others in Accessories and Bags as Gender Neutral while leaving Decor, makeup stuff as Other
df_Gucci_US.loc[(df_Gucci_US['Group'] == 'OTHER') & (df_Gucci_US['Category'].isin(['ACCESSORIES', 'BAGS', 'WATCHES'])), 'Group'] = 'GENDER NEUTRAL'

# Adding 'TRAVEL', 'GIFTS', 'OTHER' categories as OTHER GROUPS
df_Gucci_US.loc[(df_Gucci_US['Category'].isin(['TRAVEL', 'GIFTS', 'OTHER'])), 'Group'] = 'OTHER'

df_Gucci_Canada['Group'] = 'OTHER'

df_Gucci_Canada.loc[df_Gucci_Canada['category2_code'].str.contains('MEN', case=False)|df_Gucci_Canada['category1_code'].str.contains('MEN', case=False), 'Group'] = 'MEN'
df_Gucci_Canada.loc[df_Gucci_Canada['category2_code'].str.contains('WOMEN', case=False)|df_Gucci_Canada['category1_code'].str.contains('MEN', case=False), 'Group'] = 'WOMEN'
df_Gucci_Canada.loc[df_Gucci_Canada['category2_code'].str.contains('BOYS', case=False), 'Group'] = 'BOYS'
df_Gucci_Canada.loc[df_Gucci_Canada['category2_code'].str.contains('GIRLS', case=False), 'Group'] = 'GIRLS'
df_Gucci_Canada.loc[df_Gucci_Canada['category2_code'].str.contains(r'BABY|BABIES|TODDLER', case=False), 'Group'] = 'KIDS'

# Adding others in Accessories and Bags as Gender Neutral while leaving Decor, makeup stuff as Other
df_Gucci_Canada.loc[(df_Gucci_Canada['Group'] == 'OTHER') & (df_Gucci_Canada['Category'].isin(['ACCESSORIES', 'BAGS', 'WATCHES'])), 'Group'] = 'GENDER NEUTRAL'

# Adding 'TRAVEL', 'GIFTS', 'OTHER' categories as OTHER GROUPS
df_Gucci_Canada.loc[(df_Gucci_Canada['Category'].isin(['TRAVEL', 'GIFTS', 'OTHER'])), 'Group'] = 'OTHER'

# Adding others in KIDS as KIDS Group
df_Gucci_Canada.loc[(df_Gucci_Canada['Group'] == 'OTHER') & (df_Gucci_Canada['Category'].isin(['KIDS'])), 'Group'] = 'KIDS'
```

```python
#count the records in each group
group_count_Gucci_US = df_Gucci_US['Group'].value_counts()
print(group_count_Gucci_US)

group_count_Gucci_Canada = df_Gucci_Canada['Group'].value_counts()
print(group_count_Gucci_Canada)

```

```python
group_other_combinations_US = df_Gucci_US.groupby(['Category', 'Group']).size()
print("Unique category and group combinations in US")
print(group_other_combinations_US)

group_other_combinations_Canada = df_Gucci_Canada.groupby(['Category', 'Group']).size()
print("Unique category and group combinations in Canada")
print(group_other_combinations_Canada)
```

```python
# combine both the dataframes
data_Gucci = pd.concat([df_Gucci_Canada, df_Gucci_US], ignore_index=True)
```

```python
data_Gucci.to_csv('data/df_Gucci_canada_US.csv', index=False)
```
