US Chanel

```python
df_Chanel_US
```

```python
#calculate the null values in each column of the dataframe
null_value_counts = df_Chanel_US.isnull().sum()

print("number_of_null_values:\n", null_value_counts)
```

```python
#Duplicate values check

#drop unnamed column
df_Chanel_US = df_Chanel_US.drop(columns=df_Chanel_US.columns[df_Chanel_US.columns.str.contains('Unnamed')], axis=1)

#delete duplicates
df_Chanel_US = df_Chanel_US.drop_duplicates()
print(df_Chanel_US)
```

```python
summary_statistics = df_Chanel_US.describe()
summary_statistics

#This shows price range for Chanel products is between 25 to 131000.
#The 25-75% range is 200 to 5712
```

```python
# products with over 95% price range
Q9 = df_Chanel_US['full_price'].quantile(0.95)
df_over_95_percent_price = df_Chanel_US[df_Chanel_US['full_price'] > Q9]
print("Number of products with a price over the 95%:", df_over_95_percent_price.shape[0])

#category2_code for products with price over 90% price range

category2_codes = df_over_95_percent_price['category2_code']

# Display unique category2_codes for products with prices over the 75th percentile
print("Category2 codes for products with prices over the 90th percentile:")
print(category2_codes.unique())
```

```python
#count the rows with category3_code not equal to N.A.
count = df_Chanel_US[df_Chanel_US['category3_code'] != 'N.A.'].shape[0]
count

#drop category3_code column as it is not being used
df_Chanel_US = df_Chanel_US.drop(columns=['category3_code'], axis=1)
```

```python
#calculate the date range of products listed

earliest_date = df_Chanel_US['competence_date'].min()
latest_date = df_Chanel_US['competence_date'].max()

# Display the date range
print("Date range:", earliest_date, "to", latest_date)
```

```python
#products per category
products_per_category = df_Chanel_US['category2_code'].value_counts()
df_products_count = products_per_category.rename('products_count').reset_index()

#averge price per category
average_price_per_category = df_Chanel_US.groupby('category2_code')['full_price'].mean().sort_values(ascending=False)
df_average_price = average_price_per_category.rename('average_price').reset_index()

# Merge the DataFramesn
combined_df = pd.merge(df_average_price, df_products_count, on='category2_code')

# Display the DataFrame
print("Combined DataFrame:")
print(combined_df)
```

US Gucci

```python
# checking the null values in each dataframe
# it will give us the count of null values in each column
count_null_values = df_Gucci_US.isnull().sum()
print("Number of null values")
print(count_null_values)

# Drop rows containing any null value
df_Gucci_US = df_Gucci_US.dropna()
```

```python
#dropping the unnamed: 0 column as it is duplicate
df_Gucci_US = df_Gucci_US.drop(columns=['Unnamed: 0'])

```

Check for duplicate rows

```python
duplicate_rows = df_Gucci_US.duplicated().sum()
print("Number of duplicate rows:", duplicate_rows)
```

Checking Outliers

```python
# Find the lowest value in column 'full_price'
lowest_value_Gucci_US = df_Gucci_US['full_price'].min()
print("Lowest Value Gucci Product:", lowest_value_Gucci_US)

# Find the highest value in column 'full_price'
highest_value__Gucci_US = df_Gucci_US['full_price'].max()
print("Highest Value Gucci Product:", highest_value__Gucci_US)
```

```python
# checking which products comes in lowest value range to see the outliers
lowest_value_product = df_Gucci_US[df_Gucci_US['full_price'] < 200]
lowest_value_product.groupby(['category1_code']).agg({"full_price": "mean"})
```

```python
# checking which products comes in highest value range to see the outliers
highest_value_product = df_Gucci_US[df_Gucci_US['full_price'] >= 30000]
highest_value_product.groupby(['category1_code']).agg({"full_price": "mean"})
```

```python
# checking average price per category
avg_price_per_category = df_Gucci_US.groupby(['category1_code']).agg({"full_price": "mean", "price": "count", "category2_code" : "count"})

#sorting it
avg_price_per_category_sorted = avg_price_per_category.sort_values(by='full_price', ascending=True)
avg_price_per_category_sorted
```

```python
# check which dates are present in dataset
check_dates = df_Gucci_US['competence_date'].unique()
check_dates
```

```python
# checking discount values (0 or 1)
df_discount = df_Gucci_US['flg_discount'].unique()
df_discount
```

```python
# checking category of bags

df_bags = df_Gucci_US[df_Gucci_US['category1_code'] == "HANDBAGS"]
df_bags_category = df_bags['category2_code'].unique()
print("Bags Category1:\n" , df_bags_category)

# checking category 2 of bags
df_bags_category_2 = df_bags['category3_code'].unique()
print("Bags Category2:\n" , df_bags_category_2)
```
